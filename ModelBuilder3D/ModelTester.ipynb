```python
import torch
from PIL import Image
import matplotlib.pyplot as plt
from ResNeXt101 import ResNeXt101
from AlexNet_DeepLab import AlexNetDeepLab
```

```python
model = AlexNetDeepLab()
# torch.save(model, 'G:/MachineLearning/models/SE-ResNeXt101_0.pt')
```

```python
input_batch = torch.rand(2, 3, 512, 512) - 0.5
output = model(input_batch)[0]
output = output.argmax(0)
print(output.size())

palette = torch.tensor([2 ** 25 - 1, 2 ** 15 - 1, 2 ** 21 - 1])
colors = torch.as_tensor([i for i in range(3)])[:, None] * 1.1 * palette
colors = (colors % 255).numpy().astype("uint8")

r = Image.fromarray(output.byte().cpu().numpy())
r.putpalette(colors)

plt.imshow(r)
plt.show()
```
